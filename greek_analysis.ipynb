```python
texts = [
    "Σήμερα είχε απίστευτη ζέστη 😭",
    "Δεν αντέχεται άλλο αυτή η κίνηση",
    "Το φαγητό ήταν τέλειο!!!",
    "Χαχαχα πεθαίνω 😂",
    "Τι μέρα κι αυτή...",
    "Αυτό το τραγούδι είναι φοβερό 😍",
    "Δεν μπορώ άλλο με τη δουλειά 😅",
    "Πάμε διακοπές επιτέλους",
    "Τρελή κούραση σήμερα",
    "Χαχα αυτό ήταν απίστευτο 😂"
]
```

```python
all_text = " ".join(texts)

print(all_text)
```

```python
words = all_text.split()

print(words)
```

```python
from collections import Counter

word_freq = Counter(words)

print(word_freq)
```

```python
emojis = []

for word in words:
    for char in word:
        if char in "😂😭😍😅":
            emojis.append(char)

print(emojis)
```

```python
emoji_freq = Counter(emojis)

print(emoji_freq)
```

```python
print(word_freq.most_common(10))
```

```python
import matplotlib.pyplot as plt

common_words = word_freq.most_common(5)

labels = [item[0] for item in common_words]
values = [item[1] for item in common_words]

plt.bar(labels, values)
plt.title("Most Frequent Words")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```

## Findings

The dataset shows frequent emotional expression through emojis and informal vocabulary.

Laughter markers such as "χαχα" appear repeatedly, suggesting conversational and highly informal discourse.

Emojis function as pragmatic and emotional markers in online Greek communication
